```python
import pandas as pd
import numpy as np

# Your GitHub raw URL for aapl.csv
url = 'https://raw.githubusercontent.com/Electricalelectronicsfinland/AI-in-Finance-using-Python-3-ECTS-nonstop-/main/aapl.csv'

# Read the data directly from GitHub
df = pd.read_csv(url, index_col=0)

# Drop duplicate rows
df = df.drop_duplicates()

# Locate missing values and impute with column-specific mean
for column in df.columns:
    if df[column].isnull().any():
        mean_value = df[column].mean()
        df[column] = df[column].fillna(mean_value)

# Find clear outliers and impute them
for column in df.columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace outliers with NaN
    df[column] = df[column].where((df[column] >= lower_bound) & (df[column] <= upper_bound), np.nan)

# Impute outliers with column means
for column in df.columns:
    mean_value = df[column].mean()
    df[column] = df[column].fillna(mean_value)

# Print the cleaned dataframe
print(df.round(6))
```
