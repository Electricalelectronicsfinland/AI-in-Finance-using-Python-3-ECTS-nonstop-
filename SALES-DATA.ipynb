```python
import pandas as pd

# GitHub path
url = "https://raw.githubusercontent.com/Electricalelectronicsfinland/AI-in-Finance-using-Python-3-ECTS-nonstop-/main/sales_data.csv"

# Read the CSV
df = pd.read_csv(url)

# 1. Correlation matrix of numerical columns
print(df.corr(numeric_only=True))

# 2. Total sales amount grouped by region
print(df.groupby("Region")[["Sales_Amount"]].sum())

# 3. Maximum sales amount grouped by sales representative
print(df.groupby("Sales_Rep")[["Sales_Amount"]].max())

# 4. Mean unit price grouped by product category
print(df.groupby("Product_Category")[["Unit_Price"]].mean())
```
