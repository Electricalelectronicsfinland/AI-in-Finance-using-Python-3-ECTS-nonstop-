```python
# ============================================================================
# 7.7 SAMPLING EXPLORATION - 2:1 UNDERSAMPLING (FIXED)
# ============================================================================
# Complete Program - Run this entire cell in Google Colab
# ============================================================================

# 1. INSTALL & IMPORT LIBRARIES
# ----------------------------------------------------------------------------
!pip install -q imbalanced-learn xgboost

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, confusion_matrix,
                             classification_report, roc_curve)

from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline
from collections import Counter

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")

print("="*70)
print("7.7 SAMPLING EXPLORATION - 2:1 UNDERSAMPLING (FIXED)")
print("="*70)


# 2. LOAD/CREATE IMBALANCED DATASET
# ----------------------------------------------------------------------------
print("\n[1] CREATING IMBALANCED DATASET...")
X, y = make_classification(
    n_samples=10000,
    n_features=20,
    n_informative=15,
    n_redundant=3,
    n_clusters_per_class=1,
    weights=[0.95, 0.05],
    flip_y=0.05,
    random_state=42
)

feature_names = [f'feature_{i}' for i in range(X.shape[1])]
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

print(f"Total samples: {len(df)}")
print(f"Class distribution:\n{df['target'].value_counts()}")
print(f"Imbalance ratio: {df['target'].value_counts()[0]/df['target'].value_counts()[1]:.2f}:1")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
df['target'].value_counts().plot(kind='bar', ax=ax1, color=['#1f77b4', '#ff7f0e'])
ax1.set_title('Original Class Distribution', fontsize=14, fontweight='bold')
ax1.set_xlabel('Class')
ax1.set_ylabel('Count')
ax1.set_xticklabels(['Majority (0)', 'Minority (1)'], rotation=0)
for i, v in enumerate(df['target'].value_counts().values):
    ax1.text(i, v + 50, str(v), ha='center', va='bottom', fontweight='bold')

df['target'].value_counts().plot(kind='pie', ax=ax2, autopct='%1.1f%%',
                                  colors=['#1f77b4', '#ff7f0e'], explode=(0, 0.05))
ax2.set_title('Original Class Proportions', fontsize=14, fontweight='bold')
ax2.set_ylabel('')
ax2.legend(['Majority (0)', 'Minority (1)'], loc='upper right')
plt.tight_layout()
plt.show()


# 3. TRAIN-TEST SPLIT
# ----------------------------------------------------------------------------
print("\n[2] SPLITTING DATA...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
print(f"Training class distribution: {Counter(y_train)}")
print(f"Test class distribution: {Counter(y_test)}")


# 4. APPLY 2:1 UNDERSAMPLING
# ----------------------------------------------------------------------------
print("\n[3] APPLYING 2:1 UNDERSAMPLING...")
minority_count = sum(y_train == 1)
target_majority = minority_count * 2

undersampler_2_1 = RandomUnderSampler(
    sampling_strategy={0: target_majority, 1: minority_count},
    random_state=42
)

X_train_under, y_train_under = undersampler_2_1.fit_resample(X_train, y_train)

print(f"Original training size: {len(X_train)}")
print(f"Undersampled training size: {len(X_train_under)}")
print(f"New class distribution: {Counter(y_train_under)}")
print(f"Ratio: {Counter(y_train_under)[0] / Counter(y_train_under)[1]:.2f}:1")

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
axes[0].bar(['Majority (0)', 'Minority (1)'],
            [Counter(y_train)[0], Counter(y_train)[1]],
            color=['#1f77b4', '#ff7f0e'])
axes[0].set_title('Original Training Set\n(95:5 ratio)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate([Counter(y_train)[0], Counter(y_train)[1]]):
    axes[0].text(i, v + 50, str(v), ha='center', va='bottom')

axes[1].bar(['Majority (0)', 'Minority (1)'],
            [Counter(y_train_under)[0], Counter(y_train_under)[1]],
            color=['#1f77b4', '#ff7f0e'])
axes[1].set_title('Undersampled Training Set\n(2:1 ratio)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Count')
for i, v in enumerate([Counter(y_train_under)[0], Counter(y_train_under)[1]]):
    axes[1].text(i, v + 5, str(v), ha='center', va='bottom')

comparison_df = pd.DataFrame({
    'Majority': [Counter(y_train)[0], Counter(y_train_under)[0]],
    'Minority': [Counter(y_train)[1], Counter(y_train_under)[1]]
}, index=['Original', '2:1 Undersampled'])
comparison_df.plot(kind='bar', ax=axes[2], color=['#1f77b4', '#ff7f0e'])
axes[2].set_title('Class Distribution Comparison', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Count')
axes[2].set_xlabel('Dataset')
axes[2].legend(['Majority (0)', 'Minority (1)'])
plt.tight_layout()
plt.show()


# 5. DEFINE CLASSIFIERS
# ----------------------------------------------------------------------------
classifiers = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42, max_depth=5),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100, max_depth=5),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42, n_estimators=100),
    'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
    'SVM': SVC(random_state=42, probability=True, kernel='rbf')
}

print("\n[4] CLASSIFIERS TO EVALUATE:")
for name in classifiers.keys():
    print(f"  • {name}")


# 6. TRAIN & EVALUATE ON 2:1 UNDERSAMPLED DATA
# ----------------------------------------------------------------------------
print("\n[5] TRAINING & EVALUATING CLASSIFIERS (2:1 Undersampling)...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_under)
X_test_scaled = scaler.transform(X_test)

results_2_1 = []

for name, clf in classifiers.items():
    print(f"\n  > {name}")
    clf.fit(X_train_scaled, y_train_under)
    y_pred = clf.predict(X_test_scaled)
    y_proba = clf.predict_proba(X_test_scaled)[:, 1] if hasattr(clf, 'predict_proba') else None

    results_2_1.append({
        'Classifier': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_proba) if y_proba is not None else None
    })
    print(f"    F1-Score: {results_2_1[-1]['F1-Score']:.4f}")

results_df_2_1 = pd.DataFrame(results_2_1)
print("\n[6] RESULTS - 2:1 UNDERSAMPLING:")
print(results_df_2_1.sort_values('F1-Score', ascending=False).round(4).to_string(index=False))


# 7. VISUALIZE 2:1 RESULTS
# ----------------------------------------------------------------------------
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
colors = plt.cm.Set3(np.linspace(0, 1, len(results_df_2_1)))

for idx, metric in enumerate(metrics):
    ax = axes[idx // 3, idx % 3] if idx < 3 else axes[(idx // 3), idx % 3]
    sorted_df = results_df_2_1.sort_values(metric, ascending=True)
    bars = ax.barh(sorted_df['Classifier'], sorted_df[metric], color=colors)
    ax.set_xlabel(metric)
    ax.set_title(f'{metric} Comparison (2:1)', fontsize=12, fontweight='bold')
    ax.set_xlim(0, 1.0)
    for bar, val in zip(bars, sorted_df[metric]):
        ax.text(val + 0.01, bar.get_y() + bar.get_height()/2, f'{val:.3f}',
                va='center', fontsize=9)
fig.delaxes(axes[1, 2])
plt.tight_layout()
plt.show()


# 8. COMPARE AGAINST 1:1 UNDERSAMPLING
# ----------------------------------------------------------------------------
print("\n[7] COMPARING WITH 1:1 UNDERSAMPLING...")
undersampler_1_1 = RandomUnderSampler(random_state=42)
X_train_under_1_1, y_train_under_1_1 = undersampler_1_1.fit_resample(X_train, y_train)
print(f"1:1 Distribution: {Counter(y_train_under_1_1)}")

X_train_scaled_1_1 = scaler.fit_transform(X_train_under_1_1)

results_1_1 = []
for name, clf in classifiers.items():
    clf_copy = clf.__class__(**clf.get_params())
    clf_copy.fit(X_train_scaled_1_1, y_train_under_1_1)
    y_pred = clf_copy.predict(X_test_scaled)
    y_proba = clf_copy.predict_proba(X_test_scaled)[:, 1] if hasattr(clf_copy, 'predict_proba') else None

    results_1_1.append({
        'Classifier': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_proba) if y_proba is not None else None
    })

results_df_1_1 = pd.DataFrame(results_1_1)
print("\n1:1 Results:")
print(results_df_1_1.sort_values('F1-Score', ascending=False).round(4).to_string(index=False))


# 9. COMPARISON TABLE
# ----------------------------------------------------------------------------
comparison = pd.DataFrame({
    'Classifier': results_df_2_1['Classifier'],
    'F1-2:1': results_df_2_1['F1-Score'],
    'F1-1:1': results_df_1_1['F1-Score'],
    'Difference': results_df_2_1['F1-Score'] - results_df_1_1['F1-Score']
})

print("\n[8] COMPARISON: 2:1 vs 1:1")
print(comparison.round(4).to_string(index=False))


# 10. COMPARISON VISUALIZATION
# ----------------------------------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# F1-Score Comparison
x = np.arange(len(comparison))
width = 0.35
axes[0].bar(x - width/2, comparison['F1-2:1'], width, label='2:1', color='#2ecc71')
axes[0].bar(x + width/2, comparison['F1-1:1'], width, label='1:1', color='#3498db')
axes[0].set_xlabel('Classifier')
axes[0].set_ylabel('F1-Score')
axes[0].set_title('F1-Score Comparison', fontsize=12, fontweight='bold')
axes[0].set_xticks(x)
axes[0].set_xticklabels(comparison['Classifier'], rotation=15, ha='right')
axes[0].legend()
axes[0].set_ylim(0, 1.0)
for i, v in enumerate(comparison['F1-2:1']):
    axes[0].text(i - width/2, v + 0.02, f'{v:.3f}', ha='center', fontsize=8)
for i, v in enumerate(comparison['F1-1:1']):
    axes[0].text(i + width/2, v + 0.02, f'{v:.3f}', ha='center', fontsize=8)

# Difference Plot
colors_diff = ['#e74c3c' if d < 0 else '#2ecc71' for d in comparison['Difference']]
axes[1].bar(comparison['Classifier'], comparison['Difference'], color=colors_diff)
axes[1].axhline(y=0, color='black', linestyle='-', linewidth=0.8)
axes[1].set_xlabel('Classifier')
axes[1].set_ylabel('F1 Difference (2:1 - 1:1)')
axes[1].set_title('Performance Difference', fontsize=12, fontweight='bold')
axes[1].set_xticklabels(comparison['Classifier'], rotation=15, ha='right')
for i, v in enumerate(comparison['Difference']):
    axes[1].text(i, v + 0.01, f'{v:+.3f}', ha='center', fontsize=8)

# Radar Chart - Best Classifier
best_clf = comparison.loc[comparison['F1-2:1'].idxmax(), 'Classifier']
best_2_1 = results_df_2_1[results_df_2_1['Classifier'] == best_clf].iloc[0]
best_1_1 = results_df_1_1[results_df_1_1['Classifier'] == best_clf].iloc[0]

metrics_radar = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
values_2_1 = [best_2_1[m] for m in metrics_radar]
values_1_1 = [best_1_1[m] for m in metrics_radar]

angles = np.linspace(0, 2 * np.pi, len(metrics_radar), endpoint=False).tolist()
values_2_1 += values_2_1[:1]
values_1_1 += values_1_1[:1]
angles += angles[:1]

ax = plt.subplot(1, 3, 3, projection='polar')
ax.plot(angles, values_2_1, 'o-', linewidth=2, label=f'{best_clf} (2:1)', color='#2ecc71')
ax.fill(angles, values_2_1, alpha=0.25, color='#2ecc71')
ax.plot(angles, values_1_1, 'o-', linewidth=2, label=f'{best_clf} (1:1)', color='#3498db')
ax.fill(angles, values_1_1, alpha=0.25, color='#3498db')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(metrics_radar)
ax.set_ylim(0, 1.0)
ax.set_title(f'{best_clf} - Radar Comparison', fontsize=12, fontweight='bold')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
plt.tight_layout()
plt.show()


# ============================================================================
# ============================================================================
# 11. CROSS-VALIDATION - FIXED VERSION
# ============================================================================
# ============================================================================

print("\n[9] CROSS-VALIDATION (5-Fold) - 2:1 Undersampling (FIXED)...")

# ✅ FIX: Define a function that creates a dynamic sampler for each fold
def create_dynamic_sampler(X, y):
    """Create a sampler with 2:1 ratio based on the current fold's data."""
    minority_count = sum(y == 1)
    target_majority = minority_count * 2

    # Ensure we don't request more samples than available
    actual_majority = sum(y == 0)
    if target_majority > actual_majority:
        target_majority = actual_majority
        print(f"  ⚠️ Adjusted: Using {target_majority} majority samples (max available)")

    return RandomUnderSampler(
        sampling_strategy={0: target_majority, 1: minority_count},
        random_state=42
    )

# Custom cross-validation function
cv_results = {}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, clf in classifiers.items():
    print(f"\n  > {name}")
    fold_scores = []

    for fold, (train_idx, val_idx) in enumerate(cv.split(X_train, y_train)):
        # Get fold data
        X_fold_train = X_train[train_idx]
        y_fold_train = y_train[train_idx]
        X_fold_val = X_train[val_idx]
        y_fold_val = y_train[val_idx]

        # Create dynamic sampler for this fold
        sampler = create_dynamic_sampler(X_fold_train, y_fold_train)

        # Apply undersampling
        X_resampled, y_resampled = sampler.fit_resample(X_fold_train, y_fold_train)

        # Scale
        scaler_fold = StandardScaler()
        X_resampled_scaled = scaler_fold.fit_transform(X_resampled)
        X_val_scaled = scaler_fold.transform(X_fold_val)

        # Train and evaluate
        clf_copy = clf.__class__(**clf.get_params())
        clf_copy.fit(X_resampled_scaled, y_resampled)
        y_pred = clf_copy.predict(X_val_scaled)

        # Skip fold if no positive samples in validation
        if sum(y_fold_val) == 0:
            continue

        f1 = f1_score(y_fold_val, y_pred)
        fold_scores.append(f1)

    cv_results[name] = {
        'Mean F1': np.mean(fold_scores),
        'Std F1': np.std(fold_scores),
        'Min F1': min(fold_scores),
        'Max F1': max(fold_scores),
        'Num Folds': len(fold_scores)
    }
    print(f"    Mean F1: {cv_results[name]['Mean F1']:.4f} ± {cv_results[name]['Std F1']:.4f}")

cv_df = pd.DataFrame(cv_results).T
print("\nCross-Validation Results:")
print(cv_df.round(4).to_string())

# CV Visualization
fig, ax = plt.subplots(figsize=(12, 6))
cv_df_sorted = cv_df.sort_values('Mean F1', ascending=False)
ax.barh(cv_df_sorted.index, cv_df_sorted['Mean F1'],
        xerr=cv_df_sorted['Std F1'], capsize=5, color='#3498db')
ax.set_xlabel('F1-Score')
ax.set_title('Cross-Validation Results (5-Fold) - 2:1 Undersampling', fontsize=14, fontweight='bold')
ax.set_xlim(0, 1.0)
for i, (idx, row) in enumerate(cv_df_sorted.iterrows()):
    ax.text(row['Mean F1'] + 0.01, i, f'{row["Mean F1"]:.3f} ± {row["Std F1"]:.3f}',
            va='center', fontsize=9)
plt.tight_layout()
plt.show()


# ============================================================================
# 12. FINAL SUMMARY
# ============================================================================
print("\n" + "="*70)
print("FINAL SUMMARY - 7.7 SAMPLING EXPLORATION")
print("="*70)

best_2_1 = results_df_2_1.loc[results_df_2_1['F1-Score'].idxmax()]
best_1_1 = results_df_1_1.loc[results_df_1_1['F1-Score'].idxmax()]

print(f"\n📊 Best Classifier (2:1): {best_2_1['Classifier']} (F1 = {best_2_1['F1-Score']:.4f})")
print(f"📊 Best Classifier (1:1): {best_1_1['Classifier']} (F1 = {best_1_1['F1-Score']:.4f})")

print("\n📈 Per-Classifier Comparison:")
for idx, row in comparison.iterrows():
    status = "✅ BETTER" if row['Difference'] > 0 else "❌ WORSE" if row['Difference'] < 0 else "➡️ SAME"
    print(f"  {row['Classifier']}: 2:1={row['F1-2:1']:.4f} | 1:1={row['F1-1:1']:.4f} | Δ={row['Difference']:+.4f} {status}")

avg_2_1 = results_df_2_1['F1-Score'].mean()
avg_1_1 = results_df_1_1['F1-Score'].mean()

print(f"\n📊 Average F1-Score Across All Classifiers:")
print(f"  2:1 Undersampling: {avg_2_1:.4f}")
print(f"  1:1 Undersampling: {avg_1_1:.4f}")
print(f"  Difference: {avg_2_1 - avg_1_1:+.4f}")

print("\n💡 RECOMMENDATION:")
if avg_2_1 > avg_1_1:
    print(f"  ✅ 2:1 undersampling performs better. Use {best_2_1['Classifier']} (F1={best_2_1['F1-Score']:.4f})")
else:
    print(f"  ✅ 1:1 undersampling performs better. Use {best_1_1['Classifier']} (F1={best_1_1['F1-Score']:.4f})")

print("\n🔍 KEY INSIGHTS:")
print("  • 2:1 retains more data → less information loss")
print("  • 1:1 is perfectly balanced → may overfit with small datasets")
print("  • Tree-based models (XGBoost, Random Forest) handle imbalance well")
print("  • SVM and Logistic Regression benefit more from balanced data")

print("\n📊 CROSS-VALIDATION INSIGHTS:")
print(f"  • Best CV performer: {cv_df['Mean F1'].idxmax()} (F1={cv_df['Mean F1'].max():.4f})")
print(f"  • Most stable (lowest std): {cv_df['Std F1'].idxmin()} (std={cv_df['Std F1'].min():.4f})")

print("\n" + "="*70)
print("✅ ANALYSIS COMPLETE - ALL RESULTS GENERATED")
print("="*70)


# 13. EXPORT RESULTS (Optional)
# ----------------------------------------------------------------------------
try:
    from google.colab import files

    results_df_2_1.to_csv('results_2_1_undersampling.csv', index=False)
    results_df_1_1.to_csv('results_1_1_undersampling.csv', index=False)
    comparison.to_csv('comparison_results.csv', index=False)
    cv_df.to_csv('cv_results_2_1.csv', index=True)

    print("\n📁 Results exported to CSV files:")
    print("  • results_2_1_undersampling.csv")
    print("  • results_1_1_undersampling.csv")
    print("  • comparison_results.csv")
    print("  • cv_results_2_1.csv")

    # Download
    for f in ['results_2_1_undersampling.csv', 'comparison_results.csv', 'cv_results_2_1.csv']:
        files.download(f)
    print("\n✅ Files downloaded successfully!")
except:
    print("\n⚠️ Not in Colab or download skipped - files saved locally.")
```
