```python
import pandas as pd
import numpy as np

def main():

    # Read the CSV file directly from GitHub
    url = "https://raw.githubusercontent.com/Electricalelectronicsfinland/AI-in-Finance-using-Python-3-ECTS-nonstop-/main/tsla.csv"
    df = pd.read_csv(url)

    # Remove unnecessary column
    if "Unnamed: 0" in df.columns:
        df = df.drop(columns=["Unnamed: 0"])

    # Compute 5-day rolling mean of Close
    df["Close rolling"] = df["Close"].rolling(window=5).mean()

    # Compute 5-day centered rolling mean of Volume
    df["Volume rolling"] = df["Volume"].rolling(window=5, center=True).mean()

    # Print the dataframe
    print(df)

    # Find peaks in Close rolling
    peaks = []

    values = df["Close rolling"].values

    for i in range(1, len(values) - 1):
        if not np.isnan(values[i]):
            if values[i] > values[i - 1] and values[i] > values[i + 1]:
                peaks.append(i)

    # Print peak indices
    print(np.array(peaks))

    return 0


if __name__ == "__main__":
    main()
```
